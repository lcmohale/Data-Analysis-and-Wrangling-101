```python
plt.figure(figsize = (20,10))
fig, ax = plt.subplots()

this = list(series.dropna())

for i in range(len(this)):
    this[i] = str(i) + "-" + this[i]

sns.barplot( x = daDF7.index,
             y = daDF7['CORP_ADM1'], hue = this)

plt.title('Computer Details for Corporate Office', fontsize = 23)
plt.xlabel('Computers Details', fontsize = 15)
plt.ylabel('Computers Details Count', fontsize = 15)

def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)



change_width(ax, .90)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
